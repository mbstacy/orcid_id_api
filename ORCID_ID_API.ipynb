```python
from orcid import orcid
```

```python
public_api= orcid.PublicAPI()
```

```python
total_records = public_api.search_public("'University of Oklahoma'")['orcid-search-results']['num-found']
```

```python
total_records
```

```python

def search_by_university(orcid_id,ringgold):
    try:
        data = public_api.read_record_public(orcid_id,'activities')
    except:
        return None
    if data['employments']:
        #print(data['employments'])
        for empl in data['employments']['employment-summary']:
            if empl['organization']:
                try:
                    data_ringgold = empl['organization']['disambiguated-organization']['disambiguated-organization-identifier']
                    if data_ringgold == ringgold and empl['end-date'] == None:
                        return empl
                    else:
                        return None
                except:
                    return None
```

```python
current_ou_people = []
for page in range(0,total_records,100)[:10]:
    for person in public_api.search_public("'University of Oklahoma'",start=page, rows=100)['orcid-search-results']['orcid-search-result']:
        current_person = search_by_university(person['orcid-profile']['orcid-identifier']['path'],'6187')
        if current_person:
            temp = {}
            try:
                temp['fname'] = person['orcid-profile']['orcid-bio']['personal-details']['given-names']['value']
            except:
                temp['fname']=''
            try:
                temp['lname'] = person['orcid-profile']['orcid-bio']['personal-details']['family-name']['value']
            except:
                temp['lname']= ''
            temp['name'] = "%s %s" % (temp['fname'],temp['lname'])
            temp['orcid-identifier'] = person['orcid-profile']['orcid-identifier']
            temp['employment'] = current_person
            current_ou_people.append(temp)
```
